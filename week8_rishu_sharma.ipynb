#  Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response


###  What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling



```python

# Calculator

def calculator(expression: str) -> str:
    """
    Evaluate a mathematical expression.

    Parameters:
        expression (str): Mathematical expression (e.g., "5+10*2")

    Returns:
        str: Result of the calculation or an error message.
    """
    try:
        result = eval(expression)
        return str(result)
    except ZeroDivisionError:
        return "Error: Division by zero."
    except SyntaxError:
        return "Error: Invalid mathematical expression."
    except Exception as e:
        return f"Error: {e}"
```

```python

#   Keyword Extractor


import string

def extract_keywords(text: str) -> list:
    """
    Extract keywords from the given text.

    Parameters:
        text (str): Input sentence.

    Returns:
        list: List of unique keywords.
    """
    try:
        # Convert to lowercase
        text = text.lower()

        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Split into words
        words = text.split()

        # Basic stop words
        stop_words = {
            "the", "is", "are", "and", "for", "with", "this", "that",
            "from", "have", "has", "will", "into", "your", "their",
            "about", "been", "were", "was", "than", "then", "they",
            "them", "also", "very", "over", "under", "what", "when",
            "where", "which", "while", "whose", "would", "could",
            "should", "after", "before", "there", "here", "these",
            "those", "you", "your", "ours", "our"
        }

        # Extract unique keywords
        keywords = []
        for word in words:
            if len(word) > 4 and word not in stop_words:
                if word not in keywords:
                    keywords.append(word)

        return keywords[:5]

    except Exception:
        return []
```

##  Implement Agent Logic Below

 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

```python

#  AGENT FUNCTION


def agent(query: str):
    """
    Routes the user query to the appropriate tool
    (Calculator, Keyword Extractor, or General Handler).
    """

    query_lower = query.lower().strip()

    # Calculator Route
    if "calculate" in query_lower:

        # Remove the word 'calculate'
        expression = query_lower.replace("calculate", "").strip()

        result = calculator(expression)

        return {
            "type": "calculator",
            "query": query,
            "result": result
        }

    # Keyword Extraction Route
    elif "keywords" in query_lower:

        # Remove the word 'keywords'
        text = query.replace("keywords", "").replace("Keywords", "").strip()

        result = extract_keywords(text)

        return {
            "type": "keywords",
            "query": query,
            "result": result
        }

    # General Conversation Route
    else:

        return {
            "type": "general",
            "query": query,
            "result": "General conversational input detected. No specific tool invoked."
        }
```

```python

#  TEST CASES


queries = [
    "Calculate 20 + 5",
    "keywords Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate (15 * 4) + 10",
    "keywords Python is one of the most popular programming languages",
    "Hello, how are you?"
]

for i, q in enumerate(queries, start=1):
    print(f"Test Case {i}")
    print("Query   :", q)

    response = agent(q)

    print("Response:", response)
    print("-" * 60)
```

```python
#  INTERACTIVE MODE

print("=" * 50)
print(" AI Agent Started")
print("Type your query below.")
print("Type 'exit' to stop the program.")
print("=" * 50)

while True:
    user_input = input("\nEnter query: ").strip()

    if user_input.lower() == "exit":
        print("\n Thank you for using the AI Agent!")
        break

    try:
        response = agent(user_input)

        print("\nResponse:")
        print(response)

    except Exception as e:
        print(f"\nError: {e}")
```
